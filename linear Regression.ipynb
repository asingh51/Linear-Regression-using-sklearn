# Linear Regression- Sklearn

```python
import numpy as np
import pandas as pd
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("HouseData.csv")
sqft_living=dataset['sqft_living']
price=dataset['price']
x = np.array(sqft_living).reshape(-1, 1)
y = np.array(price)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=1/3, random_state=0)
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
ypred = regressor.predict(xtest)
```

```python
plt.scatter(xtrain, ytrain, color= 'green')
predicted= regressor.predict(xtrain)
plt.plot(xtrain, predicted, color = 'blue')
plt.title ("Training Dataset Visualization")
plt.xlabel("sqft_living")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(xtest, ytest, color= 'green')
predicted= regressor.predict(xtest)
plt.plot(xtest, predicted, color = 'blue')
plt.title("Test DataSet Visualization")
plt.xlabel("sqft_living")
plt.ylabel("Price")
plt.show()
```

```python
#Performance: Returns the coefficient of determination R^2 of the prediction.

print(regressor.score(xtrain, ytrain))

print("The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a R^2 score of 0.0.")
```
